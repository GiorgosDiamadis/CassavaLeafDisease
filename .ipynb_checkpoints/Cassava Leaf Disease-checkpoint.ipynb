```python
import numpy as np
import pandas as pd
import tensorflow as tf
from tensorflow import keras
import matplotlib.pyplot as plt
import imblearn
import cv2
from tensorflow.keras.optimizers import *
from sklearn.metrics import accuracy_score, classification_report
from tensorflow.keras.preprocessing.image import ImageDataGenerator
import os
from pyimagesearch.clr_callback import CyclicLR
from pyimagesearch import config
import matplotlib.pyplot as plt
from pyimagesearch.learningratefinder import LearningRateFinder
```

```python
def prepare_tf():
    config = tf.compat.v1.ConfigProto()
    config.gpu_options.allow_growth = True

    return tf.compat.v1.Session(config=config)

def process(bgr):
    bgr = bgr
    lab = cv2.cvtColor(bgr, cv2.COLOR_BGR2LAB)

    lab_planes = cv2.split(lab)

    clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(12, 12))

    lab_planes[0] = clahe.apply(lab_planes[0])

    lab = cv2.merge(lab_planes)

    bgr = cv2.cvtColor(lab, cv2.COLOR_LAB2BGR)
    
    
    
    return bgr
prepare_tf()
```

```python
class CassavaClassifier:
    def __init__(self,data_path):
        self.target_size = None
        self.batch_size = None
        self.learning_rate = None
        self.model = None
        self.optimizer = None
        self.dir = data_path
        
    def create_generators(self,target_size,batch_size):
        
        self.target_size = target_size
        self.batch_size = batch_size
        
        train_datagen = ImageDataGenerator(validation_split=0.2,
#                                            preprocessing_function=process,
#                                            rotation_range = 45,zoom_range = 0.2,
#                                            horizontal_flip = True,vertical_flip = True,
#                                            fill_mode = 'nearest',shear_range = 0.1,
#                                            height_shift_range = 0.1,width_shift_range = 0.1,
                                           rescale=1/255)
        
        self.train_generator = train_datagen.flow_from_directory(
                                 directory = self.dir,
                                 subset = "training",
                                 target_size = (target_size, target_size),
                                 batch_size = batch_size,
                                 class_mode = "sparse")

        valid_datagen = ImageDataGenerator(validation_split=0.1,
                                           preprocessing_function=process,
                                           rescale=1/255)
        self.valid_generator = train_datagen.flow_from_directory(
                                 directory = self.dir,
                                 subset = "validation",
                                 target_size = (target_size, target_size),
                                 batch_size = batch_size,
                                 class_mode = "sparse")
        
        
    def create_model(self,optimizer):
        self.optimizer = optimizer
        m = tf.keras.Sequential([
        tf.keras.layers.Conv2D(32, (5, 5),padding='same', activation='relu', input_shape=(self.target_size,
                                                                               self.target_size,
                                                                               3)),
        tf.keras.layers.MaxPooling2D(3, 3),
        tf.keras.layers.BatchNormalization(),

        tf.keras.layers.Conv2D(64, kernel_size=(3, 3),padding='same', activation='relu'),

        tf.keras.layers.MaxPooling2D(3, 3),
        tf.keras.layers.BatchNormalization(),

        tf.keras.layers.Conv2D(128, (3, 3), padding='same',activation='relu'),

        tf.keras.layers.MaxPooling2D(3, 3),
        tf.keras.layers.BatchNormalization(),

        tf.keras.layers.GlobalAveragePooling2D(),

        tf.keras.layers.Dense(units=1024, activation='relu'),
        tf.keras.layers.Dense(units=512, activation='relu'),
        tf.keras.layers.Dense(units=256, activation='relu'),
        tf.keras.layers.Dense(units=5, activation='softmax')
        ])

        m.compile(optimizer=self.optimizer,
                      loss='sparse_categorical_crossentropy',
                      metrics=['accuracy'])
        self.model = m
        
    def fit(self,epochs,save_weights,save_history,lr_scheduler=None):
        mcp_save = tf.keras.callbacks.ModelCheckpoint(save_weights,
                                                          save_best_only=True,
                                                          monitor='val_accuracy',
                                                          mode='max',
                                                          verbose=0)
        
        callbacks = [mcp_save]
        if lr_scheduler is not None:
            callbacks.append(lr_scheduler)
        
        hist = self.model.fit(self.train_generator,
                                  steps_per_epoch=len(self.train_generator),
                             validation_data=self.valid_generator,
                             validation_steps=len(self.valid_generator),
                             callbacks=callbacks,epochs=epochs)

        model_json = self.model.to_json()
        with open("model_json.json", "w") as json_file:
                json_file.write(model_json)

        hist_df = pd.DataFrame(hist.history)
        hist_csv_file = save_history
        with open(hist_csv_file, mode='w') as f:
            hist_df.to_csv(f)
        
    def load_model(self,model,weights):
        json_model = open("model_json.json", "r")
        loaded_model_js = json_model.read()
        json_model.close()
        m = tf.keras.models.model_from_json(loaded_model_js)
        m.load_weights(weights)

        m.compile(optimizer=self.optimizer,
                      loss='sparse_categorical_crossentropy',
                      metrics=tf.keras.metrics.AUC())
            
        self.model = m
  
    def predict(self):
        pass
```

# Dealing with class imbalance issue to increase accuracy

## Trying target size 128

```python
from imblearn.over_sampling import SMOTE
from tensorflow import keras
from sklearn.model_selection import train_test_split


def oversample(dim):
    train_path = "train_images/"
    classes = os.listdir(train_path)

    images = []
    labels = []

    for cl in classes:
        class_path = os.path.join(train_path, cl)
        for image in os.listdir(class_path):
            image_path = os.path.join(class_path, image)

            img = cv2.imread(image_path)
            img = cv2.resize(img, (dim, dim))
            img = process(img)

            images.append(img)
            labels.append(cl)
            
    images = np.array(images)
    labels = np.array(labels)
    return images, labels


dimension = 128
i = 0
train_images, train_labels = oversample(dim=dimension)

s = SMOTE()

train_images = train_images.reshape(train_images.shape[0], -1)

x, y = s.fit_sample(train_images, train_labels)

x = x.reshape(y.shape[0], 128, 128, 3)

for img,lab in zip(x,y):
    cv2.imwrite("train_images_balanced/" + lab + "/image" + str(i) + ".jpg",img)
    i = i + 1
    
np.save("labels",train_labels)
del train_images,train_labels
```
